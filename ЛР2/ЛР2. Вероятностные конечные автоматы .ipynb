-- нормированные переходные матрицы P^[t-1,t] для моментов времени t=1,..,10;\
-- распределение вероятностей в моменты времени t=1,..,10.

```python
# подключение библиотек
import pandas as pd
import numpy as np
import itertools
import random
import math
```

```python
# функция для расчёта вероятностей переходной матрицы на соответствующем шаге 
# рассчёт тригонометрических функций производится в радианах! 
def transition_matrix(t): 
    P = []    
    P.append( [1+math.cos(t)**2, 1+math.cos(2*t), math.cos(2*t), math.cos(t)**2] )    
    P.append( [math.sin(t)**2, math.tan(t)**2, 2*math.sin(t/2)**2, math.cos(2*t)] )  
    P.append( [t**2, 1+t**2, (1+t)**2, t+t**2] )
    P.append( [math.sqrt(abs(8-3*t)), math.sqrt(abs(8-2*t)), math.sqrt(t**2-1), 2*t] )
    P = (np.array(P)).round(5)
    
    return P
```

```python
# начальное распределение вероятностей состояний вероятностного конечного автомата
P_ksi = np.array([0.1, 0.2, 0.3, 0.4])
P_ksi
```

```python
P_t = P_ksi
print("Начальное распределение вероятностей:\nP_0 =", P_t)

for t in range(1, 11):
    P_t = (np.dot(P_t, transition_matrix(t))).round(5)
    print("\nПереходная матрица в момент времени t=%i:\n" % t)
    print(transition_matrix(t))
    print("\n\n\nРаспределение вероятности в момент времени t=%i" % t) 
    print("P_%i =" % t, P_t)
```
