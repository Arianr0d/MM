```python
# подключение библиотек
import pandas as pd
import numpy as np
import itertools
import random
```

```python
#входные параметры
sigma = list([set('АБВГД'), set('01')])         # -- номера входов
ksi = list([set('01')]*3)                       # -- номера состояний
tau = list([set('01'), set('АБВГ'), set('01')]) # -- номера выходов
```

```python
print("sigma:", sigma)
print("ksi:",ksi)
print("tau",tau)
```

```python
# входы
list_input = []
print("Входы:")
for element in itertools.product(*sigma):
    list_input.append(element)
    print(element)
    
# состояния
list_condition = []
print("\nСостояния:")
for element in itertools.product(*ksi):
    list_condition.append(element)
    print(element)
    
# выходы
list_output = []
print("\nВыходы:")
for element in itertools.product(*tau):
    list_output.append(element)
    print(element)
```

## Формирование таблицы переходов для автоматов Мили и Мура

```python
# функция генерации таблицы переходов для автоматов Мили и Мура
def tb_transition(list_input, list_condition):  
    table_transition = pd.DataFrame('0', index=range(len(list_input)),
                                    columns=[ ''.join(list_condition[i]) for i in range(len(list_condition))])
    
    # заполнение таблицы случайными состояниями
    for i in table_transition.columns:
        for j in range(len(list_input)):
            table_transition[i][j] = table_transition.columns[random.randint(0,len(list_condition)-1)]
            
    table_transition['row_name'] = pd.Series([''.join(list_input[i]) for i in range(len(list_input))],
                                             index=table_transition.index)
    
    return(table_transition)
```

```python
# таблица переходов для автоматов Мили и Мура
_transition = tb_transition(list_input, list_condition)
_transition
```

## Формирование таблицы выходов для автомата Мили

```python
# функция генерации таблицы выходов для автомата Мили
def tb_output(list_input, list_condition, list_output):
    table_output = pd.DataFrame('0', index=range(len(list_input)),
                                    columns=[ ''.join(list_condition[i]) for i in range(len(list_condition))])
    
    # заполнение таблицы случайными выходами
    for i in table_output.columns:
        for j in range(len(list_input)):
            table_output[i][j] = ''.join(list_output[random.randint(0,len(list_output)-1)])
            
    table_output['row_name'] = pd.Series([''.join(list_input[i]) for i in range(len(list_input))],
                                             index=table_output.index)
    
    return(table_output)
```

```python
# таблица выходов для автомата Мили
_output = tb_output(list_input, list_condition, list_output)
_output
```

## Формирование отмеченной таблицы переходов для автомата Мура

```python
# функция для составления отмеченной таблицы переходов для автомата Мура
def tb_marked(_transition, list_output):
    table_marked = _transition.copy(deep=True)
    row = []
    
    [row.append(''.join(list_output[random.randint(0,len(list_output)-1)]))
                    for i in range(len(_transition.columns)-1)]
    row.append('y1y2y3')
    table_marked.loc[_transition.shape[0]] = row
    
    return(table_marked)
```

```python
# отмеченная таблица переходов для автомата Мура
_marked = tb_marked(_transition, list_output)
_marked
```

## Формирование таблицы выходов для интерпретирующего автомата Мили

```python
# функция для составления таблицы выходов интерпретирующего автомата Мили
def tb_interpreter_out(_transition, _output, _marked):
    table_interpreter_out = _output.copy(deep=True)
    
    for i in _transition.columns[0:_transition.columns.shape[0]-1]:
        for j in range(_transition.shape[0]):
            if _output[i][j] == _marked[str(_transition[i][j])][_marked.shape[0]-1]:
                continue
            else: 
                table_interpreter_out[i][j] = 'x'
    
    return(table_interpreter_out)
```

```python
# таблица выходов интерпретирующего автомата Мили
tb_interpreter_out(_transition, _output, _marked)
```

## Формирование матрицы соединений конечного автомата

```python
# функция для составления матрицы соединений конечного автомата
def tb_compound(_transition, _output):
    table_compound = pd.DataFrame('0', index=range(_transition.shape[1]-1),
                                    columns=_transition.columns)
    for i in range(table_compound.shape[0]):
        table_compound['row_name'][i] = table_compound.columns[i]
        
    for i in table_compound.columns[0:table_compound.columns.shape[0]-1]:
        for j in range(table_compound.shape[0]):
            cell = _transition.index[_transition[i] == 
                        table_compound[ table_compound.columns[table_compound.shape[1]-1] ][j]]
            if len(cell) != 0:
                table_compound[i][j] = '';
                for k in range(len(cell)):
                    table_compound[i][j] += _output[_output.columns[_output.shape[1]-1]][cell[k]] + \
                            '/' + _output[i][cell[k]]
                    if k != len(cell)-1:
                        table_compound[i][j] += ' v '
    
    return(table_compound)
```

```python
# матрица соединений конечного автомата
tb_compound(_transition, _output)
```
